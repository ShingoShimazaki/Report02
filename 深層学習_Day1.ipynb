## <b>Section2: 活性化関数

<b>要点まとめ

<b>活性化関数は次の層への出力の大きさを決める非線形関数であり、代表的なものとしてステップ関数、シグモイド関数、ReLU関数などがある。勾配消失問題の回避やスパース化等の理由から、今日ではReLU関数が広く用いられている。

* 活性化関数
 * ニューラルネットワークにおいて、次の層への出力の大きさを決める<u>非線形の関数</u>。
 * 入力値によって、次の層への信号のON/OFFや強弱を定める働きをもつ。

#### <b>活性化関数: ステップ関数

* 数式

$$
f(x) = \begin{cases}
    1 & (x \geqq 0) \\
    0 & (x < 0)
  \end{cases}
$$

 * サンプルコード

```python
import numpy as np
import matplotlib.pylab as plt

#ステップ関数の定義
def step_function(x):
    return np.array(x > 0, dtype=np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)

plt.plot(x,y)
plt.show()
```

* 特徴
 * しきい値を超えたら発火する関数であり、出力は常に1か0。パーセプトロン（ニューラルネットワークの前身）で利用された関数。
<br><br>
* 課題
 * 0 -1間の間を表現できず、線形分離可能なものしか学習できなかった。

#### <b>活性化関数: シグモイド関数

* 数式

$$
f(u) = \frac{1}{1+e^{-u}}
$$

 * サンプルコード

```python
import numpy as np
import matplotlib.pylab as plt

#シグモイド関数の定義
def sigmaid(x):
    return 1/(1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmaid(x)

plt.plot(x,y)
plt.show()
```

* 特徴
 * 0 ~ 1の間を緩やかに変化する関数で、ステップ関数ではON/OFFしかない状態に対し、信号の強弱を伝えられるようになり、予想ニューラルネットワーク普及のきっかけとなった。
<br><br>
* 課題
 * 大きな値では出力の変化が微小なため、勾配消失問題を引き起こす事があった。

#### <b>活性化関数: ReLU関数

* 数式

$$
f(x) = \begin{cases}
    x & (x > 0) \\
    0 & (x \leqq 0)
  \end{cases}
$$

* サンプルコード

```python
import numpy as np
import matplotlib.pylab as plt

#ReLU関数の定義
def relu(x):
    return np.maximum(0,x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x,y)
plt.show()
```

* 特徴
 * 今最も使われている活性化関数
 * <u>勾配消失問題</u>の回避と<u>スパース化</u>に貢献することで良い成果をもたらしている。

## <b>Section3: 出力層

<b>要点まとめ

<b>出力層より出力された値について、正解値とのギャップを誤差関数により定義し、フィードバックをかけることになる。回帰タスクでは恒等写像、二値分類ではシグモイド関数、他クラス分類ではソフトマックス関数と、タスクごとに用いるべき誤差関数は決まっている。出力層では中間層と異なり、信号の大きさはそのままに値を扱いやすい形へ変換することになる。

### <b>3-1 誤差関数

* 誤差関数
 * 出力層から出力された結果と訓練データ（正解値）とのギャップ（二乗誤差）

$$
E_{n}(w) = \frac{1}{2} \sum_{j=1}^{J} (y_{j} - d_{j})^{2} = \frac{1}{2} ||(y - d)||^{2}
$$

### <b>確認テスト8

#### <b>【問題】

誤差関数はなぜ引き算でなく二乗するか述べよ。また、誤差関数の1/2はどういう意味を持つか述べよ。

#### <b>【解答】

単純に差分の和を取ってしまうとプラスマイナスが混在し正しい評価できなくなってしまう。ここでは正解値と推定値にどの程のギャップがあったかを知りたいため、二乗することによりすべてが正の値となるように整理している。また、1/2については誤差関数を微分した時に累乗の2を打ち消すようにする計算の便宜上の意味がある。

### <b>確認テスト9

#### <b>【問題】

ソフトマックス関数の左辺、右辺分子、右辺分母のそれぞれに該当するソースコードを示し、1行ずつ処理の説明をせよ。

```python
def softmax(x):
    if x.ndim == 2:
        x = x.T
        x = x - np.max(x, axis=0)
        y = np.exp(x) / np.sum(np.exp(x), axis=0)
        return y.T
    x = x - np.max(x) #オーバーフロー対策
    return np.exp(x) / np.sum(np.exp(x))
```

#### <b>【解答】

左辺：softmax(x):　・・・　関数定義。<br>
右辺分子：np.exp(x)　・・・　xの指数を取っている。<br>
右辺分母：np.sum(np.exp(x))　・・・　xの指数の和を取っている。

#### <b>「確認テスト9」について自身の考察結果

<b>ソフトマックス関数の式自体を表しているのは"np.exp(x) / np.sum(np.exp(x))"の一行のみであるが、実際にはデータの型式や行列形状を整えプログラムを安定させるための処理が必要となる。この処理が不十分であるとエラーの原因となりがちであるため注意が必要となる。

### <b>確認テスト10

交差エントロピーの左辺、右辺それぞれに該当するソースコードを示し、1行ずつ処理の説明をせよ。

```python
def cross_entropy_error(d, y):
    if y.ndim == 1:
        d = d.reshape(1, d.size)
        y = y.reshape(1, y.size)
        
    # 教師データがone-hot-vectorの場合、正解ラベルのインデックスに変換
    if d.size == y.size:
        d = d.argmax(axis=1)
             
    batch_size = y.shape[0]
    return -np.sum(np.log(y[np.arange(batch_size), d] + 1e-7)) / batch_size
```

#### <b>「確認テスト10」について自身の考察結果

## <b>Section4: 勾配降下法

<b>要点まとめ

<b>勾配降下法が学習により予測モデルとデータとの誤差関数を最小化するため重み・バイアスパラメータの更新を図るプロセスであり、通常の勾配降下法の他に確率的勾配降下法やミニバッチ勾配降下法などの手法がある。近年は扱うデータ量も膨大であることが多いことから、一般には計算コストを抑制できるミニバッチ勾配降下法が広く用いられている。

### <b>勾配降下法

* 深層学習の目的
 * 学習を通して誤差を最小にするネットワークを作成すること
   * 誤差関数を最小化するパラメータを発見すること
     * 勾配降下法を利用してパラメータを最適化する

* 勾配降下法
 * 誤差E(w)を最小化するパラメータwを発見する
 * 学習率εの値によって学習の効率が大きく異なる（学習率が大きすぎると発散してしまい、小さすぎると収束までに時間がかかる）
 * 勾配降下法の学習率の決定、収束性向上のためのアルゴリズムについて以下の手法がよく用いられる。
   * Momentum
   * AdaGrad
   * Adadelta
   * Adam

* 数式
$$
w^{(t+1)} = w^{(t)} - \varepsilon \triangledown E
$$

$$
\triangledown E = \frac{\partial E}{\partial w} = [ \frac{\partial E}{\partial w_{1}} \cdots \frac{\partial E}{\partial w_{M}}]
$$

### <b>ミニバッチ勾配降下法

* ミニバッチ勾配降下法
 * ランダムに分割したデータの集合（ミニバッチ）Dtに属するサンプルの平均誤差を取る。
 * 確率的勾配降下法のメリットを損なわず、計算コストを低減できる（CPUを利用したスレッド並列化やGPUを利用したSIMD並列化）

* 数式
$$
w^{(t+1)} = w^{(t)} - \varepsilon \triangledown E_{t}
$$

$$
E_{t} = \frac{1}{N_{t}} \sum_{n \in D_{t}}E_{n}
$$
$$
N_{t} = |D_{t}|
$$

* 誤差勾配の計算
 * 数値微分を利用して誤差を計算する。

$$

 * そのため、Section5で紹介する誤差逆伝播法を利用する。

## <b>Section5: 誤差逆伝播法

<b>要点まとめ

<b>誤差関数の微分を順伝播だけで行っていたのでは計算量が膨大になってしまうため、計算コストを抑制するために誤差逆伝播法が用いられている。誤差逆伝播法は、算出された誤差を出力層側から順に微分し、前の層へと伝播することで最小限の計算で各パラメータでの微分値を解析的に計算する手法であり、一般に広く用いられる学習手法である。

* 誤差逆伝播法
 * 誤差勾配の計算に際し、誤差関数の微分は数値微分を用いて行われるが、各パラメータそれぞれについて都度数値微分を行っていたのでは計算負荷が大きいため、誤差逆伝播法が利用される。
 * 誤差逆伝播法は、算出された誤差を<u>出力層側から順に</u>微分し、前の層へと伝播することで最小限の計算で各パラメータでの微分値を解析的に計算する手法。計算結果（=誤差）から微分を逆算することで、不要な再帰的計算を避けて微分を算出できる。

### <b>確認テスト14

#### <b>【問題】

#### <b>【解答】

delta2 = functions.d_mean_squared_error(d, y)<br>
grad['b2'] = np.sum(delta2, axis=0)<br>
grad['W2'] = np.dot(z1.T, delta2)<br>
delta1 = np.dot(delta2, W2.T) * functions.d_sigmoid(z1)

#### <b>【問題1】

以下の【図1】のような掛け算ユニットがある計算グラフで逆伝播を行ったとき、赤枠に当てはまる数字はどれか。

#### <b>【解答1】

x0 = 100, x1 = 300, x2 = 6000, w1 = 3, w2 = 20<br>
<br>
順伝播<br>
x1 = x0 * w1 = 100 * 3 = 300<br>
x2 = x1 * w2 = 300 * 20 = 6000<br>
<br>
逆伝播<br>
dE/dx2 = 1<br>
dE/dw2 = dE/dx2・dx2/dw2 = 1 * (6000/20) = 300<br>
<br>
dE/dx1 = dE/dx2・dx2/dx1 = 1 * (6000/300) = 20<br>
<br>
dE/x0 = dE/dx2・dx2/dx1・dx1/dx0 = 1 * (6000/300) * (300/100) = 60<br>
<br>
赤枠に当てはまる数字は「60」。

#### <b>【問題2】

p(xi|C0) = 0.1, p(xi|C1) = 0.3, p(xi|C2) = 0.2, p(C0) = 0.7, p(C1) = 0.2, p(C2) = 0.1であるときxiがC1に属する確率を求めよ。

#### <b>【解答2】

p(C1|xi) = {p(xi|C1) * p(C1)} / p(xi)<br>
ここで、<br>
p(xi) = Σp(x|Ck) * p(Ck)より、
p(xi) = p(xi|C0) * p(C0) + p(xi|C1) * p(C1) + p(xi|C2) * p(C2) = 0.1 * 0.7 + 0.3 * 0.2 + 0.2 * 0.1 = 0.15<br>
であるから、求める確率はp(C1|xi)は<br>
p(C1|xi) = {0.3 * 0.2} / 0.15 = 0.4
