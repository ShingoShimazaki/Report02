# <b>深層学習: Day4

## <b>Section1: 強化学習

<b>要点まとめ

<b>強化学習は「行動を学習する仕組み」であり、ある環境下で目的とする報酬を最大化するためにはどのような行動をとっていけばいいかを学習する仕組みのこと。強化学習は一連の行動系列の結果としての報酬を最大とするように学習するところが教師あり学習とは大きく異なる。

* <b>【参考文献】
    * ディープラーニングG検定公式テキスト, 日本ディープラーニング協会（2018）, p174

### <b>1-1 強化学習とは

* 強化学習
 * 長期的に報酬を最大化できるように環境のなかで行動を選択できるエージェントを作ることを目標とする機械学習の一分野
 * 行動の結果として与えられる利益（報酬）をもとに、行動を決定する原理を改善していく仕組み

### <b>1-2 強化学習の応用例

* マーケティングの場合
 * 環境: 会社の販売促進部
 * エージェント: プロフィールと購入履歴に基づいて、キャンペーンメールを送る顧客を決めるソフトウェア
 * 行動: 顧客ごとに送信、非送信のふたつの行動を選ぶ
 * 報酬: キャンペーンのコストという負の報酬とキャンペーンで生み出されると推測される売上という正の報酬を受ける

### <b>1-3 探索と利用のトレードオフ

* 環境について事前に完璧な知識があれば、最適な行動を予測し決定することは可能。
* 強化学習の場合、上記仮定は成り立たないとする。不完全な知識を元に行動しながら、データを収集。最適な行動を見つけていく。
<br><br>
* 過去のデータで、ベストとされる行動のみを常に取り続ければ他にもっとベストな行動を見つけることはできない（探索不足）。
* 未知の行動のみを常に取り続ければ、過去の経験が活かせない（利用不足）。

### <b>1-4 強化学習のイメージ

* <b>【参考文献】
    * 田島 太一, オンライン型深層強化学習を用いた金融商品のポートフォリオ管理戦略手法の提案, 早稲田大学 2017年度修士論文, p10, 図3.2.1

### <b>1-5 強化学習の差分

* 教師あり・なし学習と強化学習の違い
 * 教師なし、あり学習では、データに含まれるパターンを見つけ出すおよびそのデータから予測することが目標
 * 強化学習では、優れた方策を見つけることが目標

* 強化学習の歴史
 * 冬の時代があったが、計算速度の進展により大規模な状態をもつ場合の、強化学習を可能としつつある。
 * 関数近似法と、Q学習を組み合わせる手法の登場
* Q学習
 * 行動価値関数を、行動する毎に更新することにより学習を進める方法
* 関数近似法
 * 価値関数や方策関数を関数近似する手法のこと

### <b>1-6 行動価値関数

* 行動価値関数
 * 価値を表す関数としては、状態価値関数と行動価値関数の2種類がある
 * ある状態の価値に注目する場合は、状態価値関数
 * 状態と価値を組み合わせた価値に注目する場合は、行動価値関数

### <b>1-7 方策関数

* 方策関数
 * 方策ベースの強化学習手法において、ある状態でどのような行動を採るのかの確率を与える関数

### <b>1-8 方策勾配法

* 方策反復法
 * 方策をモデル化して最適化する手法<br><br>
 $$
 \theta^{(t+1)} = \theta^{(t)} + \epsilon \nabla J (\theta)
 $$

   * J: 方策の良さ（要定義）


* 定義方法
   * 平均報酬
   * 割引報酬和
 * 上記の定義に対応して、行動価値関数:Q(s,a)の定義を行い。方策勾配定理が成り立つ。<br><br>
 $$
 \nabla_{\theta} J (\theta) = \mathbb{E}_{\pi_{\theta}} [(\nabla_{\theta}log \pi_{\theta}(a|s)Q^{\pi}(s,a))]
 $$
 <br>
 * 方策勾配定理は基本的には下記の2式から導出される
   * 状態価値関数
     * v(s) = sum_a(π(a|s)Q(s,a))
   * ベルマン方程式
     * Q(s,a) = sum_s'(P(s'|s,a))[r(s,a,s') + γV(s')]

## <b>Section2: Alpha Go

<b>要点まとめ

<b>Alpha Goはディープラーニングが注目されるきっかけにもなった、DeepMind社が開発した囲碁のプログラム。Alpha Goは状態や行動の評価にCNNを用いている、また、実際にどのような手を打つべきかの探索にはモンテカルロ木探索が用いられている。Alpha GoにはAlphaGo Leeと呼ばれるものとAlphaGo Zeroと呼ばれるものがあり、前者は囲碁の棋譜を学習に利用していたが、後者は完全に自己対局のみで学習していくプログラムとなっている。

* <b>【参考文献】
    * ディープラーニングG検定公式テキスト, 日本ディープラーニング協会（2018）, p176

* AlphaGo Lee
 * PolicyNet（方策関数）
 * ValueNet（価値関数）
 * RollOutPolicy（ニューラルネットではなく線形の方策関数）
<br><br>
* AlphaGo Leeの学習ステップ
 1. 教師あり学習によるRollOutPolicyとPolicyNetの学習
 2. 強化学習によるPolicyNetの学習
 3. 強化学習によるValueNetの学習
<br><br>
* PolicyNetの教師あり学習
 * 過去の棋譜データ（人間 対 人間のデータ）から３０００万局面分の教師を用意し、教師と同じ着手を予測できるよう学習を行った。
* PolicyNetの強化学習
 * 現状のPolicyNetとPolicyPoolからランダムに選択されたPolicyNetと対局シミュレーションを行い、その結果を用いて方策勾配法で学習。
* ValueNetの学習
 * PolicyNetを使用して対局シミュレーションを行い、その勝敗結果を教師として学習

* AlphaGo LeeとAlphaGo Zeroの違い
 1. 教師あり学習を一切行わず、強化学習のみで作成
 2. 特徴入力からヒューリスティックな要素を排除し、石の配置のみにした
 3. PolicyNetとValueNetを１つのネットワークに統合した
 4. Residual Net（後述）を導入した
 5. モンテカルロ木探索からRollOutシミュレーションをなくした
<br><br>
* AlphaGo Zeroの学習法
 1. 自己対局による教師データの作成（現状のネットワークでモンテカルロ木探索を用いて自己対局）
 2. 学習（自己対局で作成した教師データを使い学習）
 3. ネットワークの更新（学習後、現状のネットワークと学習後のネットワークとで対局テストを行い、学習後のネットワークの勝率が高かった場合、学習後のネットワークを現状のネットワークとする）

## <b>Section3: 軽量化・高速化技術

<b>要点まとめ

<b>軽量化・高速化技術は商業的な面において特に重要な要素となる。モデル並列化・データ並列化・GPUなどに代表されるモデルを如何に早く学習させるか（高速化）という技術と、量子化・蒸留・プルーニングなどに代表される手頃な（スマホなど）計算機でも動かせるモデルを構築するか（軽量化）の2つに大別される。

* 分散深層学習
 * 深層学習は多くのデータを使用したり、パラメータ調整のために多くの時間を使用したりするため、高速な計算が求められる（毎年10倍のペースで必要な計算量が増えると言われている）。
 * 複数の計算資源（ワーカー）を使用し、並列的にニューラルネットを構成することで、効率の良い学習を行いたい。
 * データ並列化、モデル並列化、GPUによる高速技術は不可欠である。

### <b>3-1 データ並列化

<b>要点まとめ

<b>演算装置をかき集めて並列に学習させる手法（世界中のスマホの空き時間を有効活用するなど！）。もっとも一般的な高速化技術の一つ。各ワーカー（演算装置）にタスクを分散させて計算を行う。同期型（各ワーカーが計算を終えるのを待つタイプ）と非同期型（各ワーカーがお互いの計算を待たないタイプ）がある。

* 演算装置をかき集めて並列に学習させる手法（世界中のスマホの空き時間を有効活用するなど！）
* 親モデルを各ワーカーに子モデルとしてコピーし、データを分割して各ワーカーごとに計算させる
<br><br>
* 同期型
 * 各ワーカーが計算が終わるのを待ち、全ワーカーの勾配が出たところで勾配の平均を計算し、親モデルのパラメータを更新する。
* 非同期型
 * 各ワーカーはお互いの計算を待たず、各子モデルごとに更新を行う。学習が終わった子モデルはパラメータサーバにPushされる。新たに学習を始める時は、パラメータサーバからPopしたモデルに対して学習していく。
* 同期型と非同期型の比較
 * 処理のスピードは、お互いのワーカーの計算を待たない非同期型の方が早い。
 * 非同期型は最新のモデルのパラメータを利用できないので、学習が不安定になりやすい。
 * 現在は同期型の方が精度が良いことが多いので、主流となっている。

* <b>【参考文献】
    * [分散深層学習とモデル並列性](https://tech.preferred.jp/ja/blog/model-parallelism-in-dnn/), 2020.12.21 アクセス

### <b>3-2 モデル並列化

<b>要点まとめ

<b>学習モデルをバラけさせて並列に学習させる手法。各ワーカー（演算装置）に学習モデルを分散させて計算を行う。モデル復元は全てのデータで学習が完了したタイミングで行う。モデルサイズが大きい時はモデル並列化を行い、取り扱うデータ自体が大きい時はデータ並列化を行うと高速化が図れる。

* 学習モデルをバラけさせて並列に学習させる手法
* 親モデルを各ワーカーに分割し、それぞれのモデルを学習させる。全てのデータで学習が終わった後で、一つのモデルに復元。
* モデルが大きい時はモデル並列化を、データが大きい時はデータ並列化をすると良い。
* モデルのパラメータ数が多いほど、スピードアップの効率も向上する。

* <b>【参考文献】
    * [分散深層学習とモデル並列性](https://tech.preferred.jp/ja/blog/model-parallelism-in-dnn/), 2020.12.21 アクセス

### <b>3-3 GPU

<b>要点まとめ

<b>GPUは"graphics"の名前が表している通り、画像処理に関する演算を担うもの。ディープラーニングではテンソル（行列やベクトル）による計算が主になるため、画像処理と類似した大規模な計算処理が行われるためGPUが適している。これを画像処理以外でさらにディープラーニング用に最適化したものにGPGPU（General-Purpose computing on GPU）がある。

* <b>【参考文献】
    * ディープラーニングG検定公式テキスト, 日本ディープラーニング協会（2018）, p129

* GPGPU (General-purpose on GPU)
 * 元々の使用目的であるグラフィック以外の用途で使用されるGPUの総称
* CPU
 * 高性能なコアが少数
 * 複雑で連続的な処理が得意
* GPU
 * 比較的低性能なコアが多数
 * 簡単な並列処理が得意
 * ニューラルネットの学習は単純な行列演算が多いので、高速化が可能
<br><br>
* GPGPU開発環境
 * CUDA
   * GPU上で並列コンピューティングを行うためのプラットフォーム
   * NVIDIA社が開発しているGPUのみで使用可能。
   * Deep Learning用に提供されているので、使いやすい
 * OpenCL
   * オープンな並列コンピューティングのプラットフォーム
   * NVIDIA社以外の会社(Intel, AMD, ARMなど)のGPUからでも使用可能。
   * Deep Learning用の計算に特化しているわけではない。
   * Deep Learningフレームワーク(Tensorflow, Pytorch)内で実装されているので、使用する際は指定すれば良い

### <b>3-4 量子化

<b>要点まとめ

<b>量子化は軽量化手法の一つであり、（不必要な）精度を落とすことでメモリと演算処理の削減を図る手法。浮動小数点を64 bit（Double）から32 bit（single）に低下させるなどがしばしば行われる。精度を落とせばその分軽量化が図れるが、最低限必要な精度は確保できるように注意しなければならない。

* 量子化
 * 軽量化手法の一つ。通常のパラメータの64 bit 浮動小数点を32 bit など下位の精度に落とすことでメモリと演算処理の削減を行う。
<br><br>
* 量子化の利点
 * 計算の高速化（倍精度演算(64 bit)と単精度演算(32 bit)は演算性能が大きく違うため、量子化により精度を落とすことによりより多くの計算をすることができる）
 * 省メモリ化（ニューロンの重みを浮動小数点のbit 数を少なくし有効桁数を下げることでニューロンのメモリサイズを小さくすることができ、多くのメモリを消費するモデルのメモリ使用量を抑えることができる）
* 量子化の欠点
 * 精度の低下（64→32bitにすることで扱える小数点の有効桁が小さくなる）

* <b>【参考文献】
    * 内木 賢吾, [ディープラーニングを軽量化する「モデル圧縮」３手法](https://laboro.ai/activity/column/engineer/%E3%83%87%E3%82%A3%E3%83%BC%E3%83%97%E3%83%A9%E3%83%BC%E3%83%8B%E3%83%B3%E3%82%B0%E3%82%92%E8%BB%BD%E9%87%8F%E5%8C%96%E3%81%99%E3%82%8B%E3%83%A2%E3%83%87%E3%83%AB%E5%9C%A7%E7%B8%AE/), 2020.12.21 アクセス

### <b>3-5 蒸留

<b>要点まとめ

<b>蒸留は軽量化手法の一つであり、規模の大きなモデルの知識を使い軽量なモデルの作成を行うモデル簡略化手法の一つ。教師モデル（予測精度の高い複雑かつ大規模なモデル）と生徒モデル（教師モデルを基にした軽量化モデル）を用いて、教師モデルの重みを固定して生徒モデルの重みを更新していく手法。

* 蒸留（Distillation）
 * 規模の大きなモデルの知識を使い軽量なモデルの作成を行う（モデルの簡略化）。
 * 教師モデル
   * 予測精度の高い、複雑なモデルやアンサンブルされたモデル
 * 生徒モデル
   * 教師モデルをもとに作られる軽量なモデル
 * 教師モデルの重みを固定し生徒モデルの重みを更新していく。誤差は教師モデルと生徒モデルのそれぞれの誤差を使い重みを更新していく

* <b>【参考文献】
    * 内木 賢吾, [ディープラーニングを軽量化する「モデル圧縮」３手法](https://laboro.ai/activity/column/engineer/%E3%83%87%E3%82%A3%E3%83%BC%E3%83%97%E3%83%A9%E3%83%BC%E3%83%8B%E3%83%B3%E3%82%B0%E3%82%92%E8%BB%BD%E9%87%8F%E5%8C%96%E3%81%99%E3%82%8B%E3%83%A2%E3%83%87%E3%83%AB%E5%9C%A7%E7%B8%AE/), 2020.12.21 アクセス

### <b>3-6 プルーニング

<b>要点まとめ

<b>プルーニングはデルの精度に寄与が少ないニューロンを削減することでモデルの軽量化、高速化を図る手法。（構造解析等において、振動応答に寄与の少ない自由度を削減して簡易に解析解を得る方法に類似している）。量子化と同様に、ニューロンを削減するほどモデルの軽量化・高速化が図れるが、必要以上にニューロンを削減してしまうと本来得たい精度が確保できなくなってしまいかねないため注意が必要である。

* プルーニング（Pruning）
 * モデルの精度に寄与が少ないニューロンを削減することでモデルの軽量化、高速化を図る手法。

* <b>【参考文献】

## <b>Section4: 応用技術

### <b>4-1 MobileNet

<b>要点まとめ

<b>MobileNetはある程度の精度を保ちつつ高速・軽量に計算を行うモデル。Depthwise Convolution（フィルタサイズを1で固定。入力マップのチャネルごとに畳み込みを実施し、出力マップをそれらと結合）とPointwise Convolution（入力マップのポイントごとに畳み込みを実施）で計算量を削減している。通常の畳み込み演算ではH×W×K×K×C×M（記号は下記参照）の計算量が必要なところを、Depthwise ConvolutionによるH×W×C×K×Kの計算とPointwise ConvolutionによるH×W×C×Mの計算の和に落とし込むことができる。

* <b>【参考文献】
    * [速くて軽くて精度の良い、MobileNetのポイントをまとめてみた](https://qiita.com/simonritchie/items/f6d6196b1b0c41ca163c), 2020.12.21 アクセス

* MobileNet
 * ディープラーニングモデルは精度は良いが、その分ネットワークが深くなり計算量が増える。
 * 計算量が増えると、多くの計算リソースが必要で、お金がかかってしまう。
 * ディープラーニングモデルの軽量化・高速化・高精度化を実現

* MobileNetでは畳み込み演算を工夫している
 * 一般的な畳み込みレイヤーにおいて、入力特徴マップH×W×C, 畳み込みカーネルサイズK×K×C, 出力チャネル数Mの場合、ストライド1でパディングを適用した場合の畳み込み計算量はH×W×K×K×C×Mとなる。
 * MobileNetでは、入力マップのチャネルごとに畳み込みを実施し、出力マップをそれらと結合することで計算量をH×W×K×K×C×M（出力チャネルM分が低減されている）に削減している。

* Depthwise Separable Convolutionという手法を用いて計算量を削減している。通常の畳込みが空間方向とチャネル方向の計算を同時に行うのに対して、Depthwise Separable ConvolutionではそれらをDepthwise ConvolutionとPointwise Convolutionと呼ばれる演算によって個別に行う。
* Depthwise Convolitionはチャネル毎に空間方向へ畳み込む。すなわち、チャネル毎にDK×DK×１のサイズのフィルターをそれぞれ用いて計算を行うため、その計算量は<u>H×W×C×K×K</u>となる。
* 次にDepthwise Convolutionの出力をPointwise Convolutionによってチャネル方向に畳み込む。すなわち、出力チャネル毎に１×１×Mサイズのフィルターをそれぞれ用いて計算を行うため、その計算量は<u>H×W×C×M</u>となる。

### <b>4-2 DenseNet

<b>要点まとめ

<b>DenseNetは画像認識のための畳込みニューラルネットワークの一種。畳み込み層とプーリング層の間にDense BlockとTransition Layerという2つを交互に挟み込んでおり、これを用いることでより最適化が必要なパラメータ数を大幅に削減して学習効率の向上を図っているネットワークである。

* Dense Block
 * 出力層に前の層の入力を足しあわせる（Dense Blockのレイヤーを通り抜けるごとにチャンネルが増える）
* Transition Layer
 * CNNでは中間層でチャネルサイズを変更し、特徴マップのサイズを変更してダウンサンプリングを行うため、Transition Layerと呼ばれる層でDence blockをつないでいる
* kをネットワークのgrowth rateと呼ぶ
 * kが大きくなるほど、ネットワークが大きくなるため、小さな整数に設定するのがよい

* DenseNetとResNetの違い
 * DenseBlockでは前方の各層からの出力全てが後方の層への入力として用いられる
 * RessidualBlockでは前1層の入力のみ後方の層へ入力
* DenseNet内で使用されるDenseBlockと呼ばれるモジュールでは成⻑率(Growth Rate)と呼ばれるハイパーパラメータが存在する。
 * DenseBlock内の各ブロック毎にk個ずつ特徴マップのチャネル数が増加していく時、kを成⻑率と呼ぶ


* <b>【参考文献】

### <b>4-3 Layer正規化/Instance正規化

### <b>4-3-1 BatchNorm

<b>要点まとめ

<b>バッチ正規化は中間層ごとのデータ分布をミニバッチ単位で標準化（平均0、標準偏差1となるように正規化）する処理のこと。ニューラルネットワークにおいては学習時間の短縮や初期値依存性の低減、そして過学習の抑制などの効果がある。一方で、バッチ正規化の問題点としては、バッチサイズが小さい場合には学習が収束しなくなってしまうといったことが発生する。そのような場合に用いられるのがレイヤー正規化の処理となっている。

* <b>【参考文献】
    * ディープラーニングG検定公式テキスト, 日本ディープラーニング協会（2018）, p153-154

* バッチ正規化（Batch Norm）
 * ミニバッチに含まれるサンプルの同一チャネルが同一分布に従うように正規化
<br><br>
* レイヤー正規化（Layer Norm）
 * それぞれのサンプルのすべてのピクセルが同一分布に従うように正規化
<br><br>
* インスタンス正規化（Instance Norm）
 * さらにチャンネルも同一分布に従うように正規化

### <b>4-3-2 LayerNorm

<b>要点まとめ

<b>レイヤー正規化はそれぞれのサンプルのすべてのピクセルが同一分布に従うように正規化を施す処理のこと。特徴マップごとに正規化された特徴マップを出力しミニバッチの数に依存しないことから、学習が収束しないといった課題を回避することができる。

* <b>【参考文献】
    * ディープラーニングG検定公式テキスト, 日本ディープラーニング協会（2018）, p153-154

### <b>4-4 Wavenet

<b>要点まとめ

<b>Wavenetは音声合成と音声認識の両者を行うニューラルネットモデルのこと。大量の文章を学習したLSTMを用いると、ある文字が最初に与えられれば次にどの文字が来るかを予測し続けることにより新しい文章を生成することができ、Wavenetもこのアプローチがベースとなっている。ただ、Wavenetにおいては扱うデータは音声時系列データであるが、RNNではなく1次元の畳み込みを行うCNNを用いて音声生成を行っている。

* <b>【参考文献】
    * ディープラーニングG検定公式テキスト, 日本ディープラーニング協会（2018）, p177

* 深層学習を用いて結合確率を学習する際に、効率的に学習を行えるアーキテクチャを提案したことがWavenetの大きな貢献の1つである。提案された新しいConvolution型アーキテクチャは<u>Dilated causal convolution</u>（拡張因果関係畳み込み）と呼ばれ、結合確率を効率的に学習できるようになっている。

* Dilated causal convolutionを用いた際の大きな利点は、単純なConbolution layer（畳み込み層）に比べて<u>パラメータ数に対する受容野が広い</u>ことである。

## <b>Section5: 例題解説

### <b>5-1 Seq2seq

<b>要点まとめ

<b>Seq2Seq(sequence-to-sequence)は、EncoderとDecoderを備えたEncoder-Decoderモデルを使って系列データを別の系列データに変換するモデルのこと。自然言語分野の中でも翻訳や対話モデルの作成で主に使用される。入力の時系列データに対してEncoderが有効な情報を詰め込んだ符号へと変換し、この符号を使用してDecoderが再度変換を行い、出力となる時系列データを生成する。Auto Encoderの時系列データ版のようなイメージ。

* <b>【参考文献】
    * [PyTorchでSeq2Seqを実装してみた](https://qiita.com/m__k/items/b18756628575b177b545), 2020.12.23 アクセス

* Seq2seq
 * Encorder-Decoderモデルとも呼ばれる。言語モデルを2つ組み合わせた形となっている
 * 入力系列がEncode（隠れ状態に変換）され、その後Decode（系列に変換）する
   * 翻訳（英語→日本語）
   * 音声認識（波形→チャット）
   * チャットボット（テキスト→テキスト）
   * etc...


* 言語モデル
 * 単語の並びに対して尤度（自然さ）を評価する
   * You say goodbye → 0.092（自然）
   * You say good die → 0.00000032（不自然）


* RNN × 言語モデル
 * RNNは系列情報を内部状態に変換できる
 * 文章の各単語が現れる際の同時確率は、事後確率で分解できる
 * RNNと言語モデルを組み合わせることにより、各地点で次にどの単語が来れば自然（事後確率最大）かを評価することができる

### <b>5-2 Transformer

<b>要点まとめ

<b>Seq2seqのようなRNNベースのEncoder-Decoderモデルでは、文章が長くなると表現力が低下するという課題があった。TransformerはRNNもCNNも使わずに Attentionのみを使用した計算量や精度の改善を図っているEncoder-Decoderモデルである。再帰も畳み込みも一切使わず、かつ並列化がかなりしやすく訓練時間が圧倒的に削減されており、汎用性が高いモデルとされている。

* Attention
 * 辞書オブジェクト。検索クエリに一致するキーを検索し、対応するバリューを取り出す。

### <b>5-3 物体検知

<b>要点まとめ

<b>画像認識においては関心領域（ROI: Region of Interest）の切り出しが行われる。この切り出すときの矩形領域のことをB-Box（Bounding box）と呼ぶ。この関心領域の切り出しと切り出した領域の物体認識を行うモデルをRCNN（Regional CNN）といい、この2つのタスクを同時に行うことで高速化を図ったモデルをfast-RCNNという（さらに改良されたfaster-RCNNというモデルも存在する）。その後の発展形としてYOLO（You Only Look Once）やSSD（Single Shot Detector）等と呼ばれるモデルも存在する。

* <b>【参考文献】
    * ディープラーニングG検定公式テキスト, 日本ディープラーニング協会（2018）, p191-192

* 広義の物体認識タスク
| 　 | 出力 |
| --- | --- |
| 分類 | （画像に対し単一または複数の）クラスラベル<br>（物体の位置には関心なし） |
| 物体検知 | Bounding Box<br>（どこに何があるか） |
| 意味領域分割 | （各ピクセルに対し単一の）クラスラベル<br>（画像のどこが風船か区別） | 
| 個体領域分割 | （各プクセルに対し単一の）クラスラベル<br>（風船ごとの区別もできる） |

* 評価は分類問題と同じく混合行列が用いられる。それ以外にも以下の指標が用いられる。
 * IoU(Intersection over Union): 物体位置の予測精度も評価する
   * IoU = Area of Overlap / Area of Union
   * IoU = TP / (TP + FP + FN)

* 物体検知のフレームワーク
| 2段階検出器<br>（Two-stage detector） | 1段階検出器（One-stage detector） |
| --- | --- |
| * 候補領域の検出とクラス推定を別々に行う<br> * 相対的に精度が高い傾向<br> * 相対的に計算量が大きく推論も遅い傾向 | * 候補領域の検出とクラス推定を同時に行う<br> * 相対的に精度が低い傾向<br> * 相対的に計算量が小さく推論も速い傾向 | 

### <b>5-4 セグメンテーション

<b>要点まとめ

<b>セマンティックセグメンテーションは、R-CNNのような矩形領域を切り出すのではなくより詳細な領域分割を得るモデルのこと。入力画像のどの位置に物体が存在するのかを画素単位で指定することができれば認識精度も向上すると予想される。自動運転、ロボティックス、工業用検査、衛星画像、医療用画像処理など、詳細な領域分割を得るモデル物体認識が必要とされるところが、画像セグメンテーションの応用分野となる。

* <b>【参考文献】
    * ディープラーニングG検定公式テキスト, 日本ディープラーニング協会（2018）, p193-197

* 完全畳み込みネットワーク（FCN）
 * セマンティックセグメンテーションを実現するモデル
 * 入力画像の画素数だけ出力層が必要になるため、出力層には縦画素数×横画素数×カテゴリー数の出力ニューロンが用意される
 * 畳み込みの受容野だけ近傍の入力刺激を加えて計算するため畳み込みと反対方向の解像度を細かくする工夫が必要
 + アンサンプリングを用いて詳細な解像度を得る
